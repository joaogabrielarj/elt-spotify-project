```python
# Configura√ß√£o
CATALOGO = "spotify_analytics"
SCHEMA = "diamond"
VIEW_NAME = "vw_dm_artist"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{VIEW_NAME}"
```

### View: Dimens√£o Artist

**Prop√≥sito**: View simples que exp√µe a dimens√£o de artistas.

**O que esta view faz:**
- SELECT * FROM dm_artist
- Apenas exposi√ß√£o da tabela, sem transforma√ß√µes
- Usar no Power BI/Tableau para JOINs com fato

**Casos de uso:**
- Conectar no Power BI/Tableau
- Relacionamento com fato via cd_artist
- Filtros de artistas em dashboards

```python
query = f"""
CREATE OR REPLACE VIEW {FULL_NAME} AS
SELECT * FROM {CATALOGO}.gold.dm_artist

COMMENT 'View simples da dimens√£o artist. Fazer JOIN com fato no BI.';
"""

spark.sql(query)
print(f"‚úÖ View {FULL_NAME} criada com sucesso!")
```

### Verifica√ß√£o e Exemplos de Uso

```python
# Verifica√ß√£o
total = spark.sql(f"SELECT COUNT(*) as total FROM {FULL_NAME}").first()['total']
print(f"\nTotal de artistas: {total}")

# Amostra dos dados
print("\n=== Amostra de Dados ===")
spark.sql(f"SELECT * FROM {FULL_NAME} LIMIT 10").show(truncate=False)

print("\nüí° Nota: Esta view √© um SELECT * FROM da dimens√£o. Fa√ßa JOIN com fato no Power BI/Tableau.")
```
