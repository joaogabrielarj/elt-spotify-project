```python
# Configura√ß√£o
CATALOGO = "spotify_analytics"
SCHEMA = "gold"
VIEW_NAME = "vw_ft_track_popularity"
FULL_NAME = f"{CATALOGO}.{SCHEMA}.{VIEW_NAME}"
```

### View: Fato Track Popularity

**Prop√≥sito**: View simples que exp√µe o fato de popularidade.

**O que esta view faz:**
- SELECT * FROM ft_track_popularity
- Apenas exposi√ß√£o da tabela, sem transforma√ß√µes
- JOINs com dimens√µes devem ser feitos no Power BI/Tableau

**Casos de uso:**
- Conectar no Power BI/Tableau
- Criar relacionamentos com dimens√µes no BI
- Base para dashboards

```python
query = f"""
CREATE OR REPLACE VIEW {FULL_NAME} AS
SELECT * FROM {CATALOGO}.{SCHEMA}.ft_track_popularity

COMMENT 'View simples do fato de popularidade. Fazer JOINs com dimens√µes no BI.';
"""

spark.sql(query)
print(f"‚úÖ View {FULL_NAME} criada com sucesso!")
```

### Verifica√ß√£o e Exemplos de Uso

```python
# Verifica√ß√£o
total = spark.sql(f"SELECT COUNT(*) as total FROM {FULL_NAME}").first()['total']
print(f"\nTotal de registros na view: {total}")

# Amostra dos dados
print("\n=== Amostra de Dados ===")
spark.sql(f"SELECT * FROM {FULL_NAME} LIMIT 5").show(truncate=False)

print("\nüí° Nota: Esta view √© um SELECT * FROM do fato. Fa√ßa JOINs com dimens√µes no Power BI/Tableau.")
```
