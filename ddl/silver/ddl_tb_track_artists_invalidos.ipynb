```python
CATALOGO_DESTINO = "spotify_analytics"
SCHEMA_DESTINO = "silver"
TABELA_DESTINO = "tb_track_artists_invalidos"

nome_tabela_destino = f"{CATALOGO_DESTINO}.{SCHEMA_DESTINO}.{TABELA_DESTINO}"
```

### Configuração

Define catálogo, schema e nome da tabela de auditoria para track_artists inválidos.

### DDL - Tabela de Auditoria (Track-Artists Inválidos)

Armazena registros de relacionamento track-artist que falharam na validação:
- **Flags de validação**:
  - flag_track_id_valido
  - flag_artist_id_valido
  - flag_artist_name_valido
  - flag_qualidade

Permite identificar problemas na extração de artistas do JSON.

```python
query = f"""
CREATE OR REPLACE TABLE {nome_tabela_destino} (
    track_id STRING COMMENT 'FK para tb_tracks (pode ser nulo)',
    artist_id STRING COMMENT 'ID do artista (pode ser nulo)',
    artist_name STRING,
    dt_ingestion TIMESTAMP,
    dc_origem STRING,
    flag_track_id_valido BOOLEAN COMMENT 'Valida se track_id está preenchido',
    flag_artist_id_valido BOOLEAN COMMENT 'Valida se artist_id está preenchido',
    flag_artist_name_valido BOOLEAN COMMENT 'Valida se artist_name está preenchido',
    flag_qualidade STRING COMMENT 'Status geral: OK ou ERRO'
)
USING DELTA
COMMENT 'Tabela de auditoria: registros de track-artists que falharam na validação'
TBLPROPERTIES (
    'delta.autoOptimize.optimizeWrite' = 'true',
    'quality' = 'audit',
    'purpose' = 'data_quality_monitoring'
);
"""

spark.sql(query)
print(f"✅ Tabela {nome_tabela_destino} criada com sucesso!")
```
